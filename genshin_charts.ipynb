```python
df_data = pd.get_dummies(df_data, columns=['Element', 'Weapon', 'Role', 'Ascension Stat'], drop_first=True)
df_data['Gender']  = df_data['Gender'].str.strip().str.title().map({'Male': 0, 'Female': 1})
df_data['AAPU'] = (df_data['AAPU'].str.replace('[\\%,]', '', regex = True).astype(float)/100)
df_data['ASPU'] = (df_data['ASPU'].str.replace('[\\%,]', '', regex = True).astype(float)/100)
df_data['RPU'] = (df_data['RPU'].str.replace('[\\$,]','', regex = True)).astype(float)
df_data.columns
```

```python
df_attributes = pd.get_dummies(df_attributes, columns=['Element', 'Weapon', 'Role', 'Ascension Stat'], drop_first=True)
df_attributes['Gender']  = df_attributes['Gender'].str.strip().str.title().map({'Male': 0, 'Female': 1})
df_attributes['AAPU'] = (df_attributes['AAPU'].str.replace('[\\%,]', '', regex = True).astype(float)/100)
df_attributes['ASPU'] = (df_attributes['ASPU'].str.replace('[\\%,]', '', regex = True).astype(float)/100)
df_attributes.columns
```

```python
df_data = df_data.fillna(0)
df_attributes = df_attributes.fillna(0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# X = df_data[[ 'ATK', 'DEF', 'HP', 'AAPU', 'ASPU', 'Synergy', 'Power Score', 'Gender', 'Rerun'] + [col for col in df_data.columns if col.startswith(('Element_', 'Weapon_', 'Role_', 'Ascension Stat_'))]]
# X = df_data[['Gender', 'Rerun']+ [col for col in df_data.columns if col.startswith(('Element_', 'Weapon_', 'Role_', 'Ascension Stat_'))]]
X = df_data[['ATK', 'DEF', 'HP', 'AAPU', 'ASPU', 'Synergy', 'Power Score', 'Gender', 'Rerun'] + [col for col in df_data.columns if col.startswith(('Ascension Stat_'))]]
y = df_data['RPU']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(f"Training: {X_train.shape}")
print(f"Testing:  {X_test.shape}")

       
```

```python
from sklearn.metrics import mean_squared_error, r2_score, classification_report
from sklearn.feature_selection import RFE
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})
intercept = model.intercept_
print(coefficients.sort_values(by='Coefficient', key=abs, ascending=False))
print("Intercept:", intercept)
```

```python
scores = pd.DataFrame(data=zip(y_pred, y_test), columns=["Predicted", "Actual"])
scores["Result"] = scores["Predicted"] == scores["Actual"]
scores
```

```python
equation = f"Revenue Per User = {intercept:.2f} "
for name, coef in zip(X.columns, model.coef_):
    equation += f"+ ({model.coef_[X.columns.get_loc(name)]:.2f} * {name}) "
print("\n Final Equation:")
print(equation)
```
